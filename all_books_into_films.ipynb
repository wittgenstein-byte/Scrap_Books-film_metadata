## All books adapted into films

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
urls = [
   " https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(0%E2%80%939,_A%E2%80%93C)",
    "https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(D%E2%80%93J)",
    "https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(K%E2%80%93R)",
    "https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(S%E2%80%93Z)",
    "https://en.wikipedia.org/wiki/List_of_non-fiction_works_made_into_feature_films",
    "https://en.wikipedia.org/wiki/List_of_short_fiction_made_into_feature_films",
    "https://en.wikipedia.org/wiki/List_of_children%27s_books_made_into_feature_films"
]
```

```python
def extract_book_info(book_text):
    parts = book_text.rsplit('),', 1)  # แยกโดยใช้ '),' เพื่อแยกชื่อหนังสือกับชื่อผู้แต่ง
    title = parts[0].strip() + ')' if len(parts) > 0 else ""
    author = parts[1].strip() if len(parts) > 1 else ""
    return title, author

data = []

# วนลูปดึงข้อมูลจากทุก URL
for url in urls:
    print(f"กำลังดึงข้อมูลจาก: {url}")

    # ส่งคำขอ GET ไปยังหน้าเว็บ
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # ค้นหาตารางทั้งหมดในหน้า
    tables = soup.find_all('table', {'class': 'wikitable'})

    # วนลูปผ่านแต่ละตาราง
    for table in tables:
        df = pd.read_html(str(table))[0]

        # วนลูปผ่านแต่ละแถวใน DataFrame
        for index, row in df.iterrows():
            book_info = str(row[0])  # หนังสือ
            film_adaptations = str(row[1])  # รายชื่อหนัง

            # แยกชื่อหนังสือและผู้แต่ง
            book_title, author = extract_book_info(book_info)

            # แยกชื่อภาพยนตร์ (กรณีที่มีหลายเรื่อง)
            films = [film.strip() for film in film_adaptations.split('\n') if film.strip()]

            # สร้างแต่ละภาพยนตร์
            for film in films:
                data.append({
                    "Fictions": book_title,
                    "author": author,
                    "film_adaptation": film
                })

# JSON save
with open("all_books_into_films.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"ดึงข้อมูลสำเร็จ! รวมทั้งหมด {len(data)} รายการ บันทึกในไฟล์ 'all_books_into_films.json'")
```
