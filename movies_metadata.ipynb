## Cleaning Data

```python
import json
import ast


def safe_eval(val):
    """ แปลง string ให้เป็น dictionary หรือ list ถ้าเป็นไปได้ """
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError):
        return None

def clean_movie_data(input_file, output_file):
    # โหลดข้อมูล
    df = pd.read_csv(input_file, low_memory=False)

    cleaned_data = []
    for _, row in df.iterrows():
        belongs_to_collection = safe_eval(row["belongs_to_collection"])  # Evaluate once
        cleaned_movie = {
            # Check if belongs_to_collection is a dict before accessing 'name'
            "name": belongs_to_collection.get("name", "") if isinstance(belongs_to_collection, dict) else "",
            "genres": [genre["name"] for genre in safe_eval(row["genres"]) or []],
            "original_title": row["original_title"],
            "overview": row["overview"],
            "popularity": str(row["popularity"]),
            "release_date": row["release_date"],
            "title": row["title"],
            "vote_average": str(row["vote_average"]),
            "vote_count": str(row["vote_count"])
        }
        cleaned_data.append(cleaned_movie)

    #  JSON!
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(cleaned_data, f, ensure_ascii=False, indent=4)

# เรียกใช้งานฟังก์ชัน
input_file = "movies_metadata.csv"
output_file = "cleaned_movies_metadata.json"
clean_movie_data(input_file, output_file)
```
